```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np

# Sample data generation (replace with actual data if available)
np.random.seed(0)
data_x = np.random.uniform(0, 700000, 300)  # Easting in meters
data_y = np.random.uniform(0, 1300000, 300)  # Northing in meters
color_category = np.random.choice(['red', 'blue'], 300)  # Color category

# Define plot area for the zoomed-in map in meters
zoom_xlim = (350000, 550000)
zoom_ylim = (300000, 450000)

# Create the main figure and axis
fig = plt.figure(figsize=(10, 15))

# Top plot - full map of the UK with OSGB coordinates
ax1 = fig.add_subplot(2, 1, 1, projection=ccrs.OSGB())
ax1.set_extent([0, 700000, 0, 1300000], crs=ccrs.OSGB())  # Full UK extent in British National Grid (OSGB) coordinates
ax1.coastlines(resolution='10m')
ax1.gridlines(draw_labels=False, linewidth=0.5, color='gray', alpha=0.7, linestyle='--')

# Scatter plot for the full map
for color in ['red', 'blue']:
    ax1.scatter(data_x[color_category == color], 
                data_y[color_category == color], 
                color=color, s=10, transform=ccrs.OSGB())

# Highlight zoomed-in area with a rectangle
rect = plt.Rectangle((zoom_xlim[0], zoom_ylim[0]), 
                     zoom_xlim[1] - zoom_xlim[0], 
                     zoom_ylim[1] - zoom_ylim[0],
                     linewidth=2, edgecolor='black', facecolor='none', 
                     transform=ccrs.OSGB())
ax1.add_patch(rect)

# Bottom plot - zoomed-in section
ax2 = fig.add_subplot(2, 1, 2, projection=ccrs.OSGB())
ax2.set_extent([zoom_xlim[0], zoom_xlim[1], zoom_ylim[0], zoom_ylim[1]], crs=ccrs.OSGB())
ax2.coastlines(resolution='10m')
ax2.gridlines(draw_labels=False, linewidth=0.5, color='gray', alpha=0.7, linestyle='--')

# Zoomed-in scatter plot
for color in ['red', 'blue']:
    ax2.scatter(data_x[color_category == color], 
                data_y[color_category == color], 
                color=color, s=10, transform=ccrs.OSGB())

# Manual labels for the grid
# Top plot labels
ax1.text(-0.1, 0.5, 'Northing (m)', va='bottom', ha='center', 
         rotation='vertical', rotation_mode='anchor', transform=ax1.transAxes, fontsize=12)
ax1.text(0.5, -0.1, 'Easting (m)', va='top', ha='center', transform=ax1.transAxes, fontsize=12)

# Bottom plot labels
ax2.text(-0.1, 0.5, 'Northing (m)', va='bottom', ha='center', 
         rotation='vertical', rotation_mode='anchor', transform=ax2.transAxes, fontsize=12)
ax2.text(0.5, -0.1, 'Easting (m)', va='top', ha='center', transform=ax2.transAxes, fontsize=12)

# Add orientation arrows for "N" in each plot
for ax in [ax1, ax2]:
    ax.text(0.95, 0.9, 'N', transform=ax.transAxes, ha='center', va='center', fontsize=12)
    ax.annotate('', xy=(0.95, 0.85), xytext=(0.95, 0.95),
                arrowprops=dict(arrowstyle="->", lw=1.5), transform=ax.transAxes)

plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.patches import ConnectionPatch
import cartopy.crs as ccrs
import numpy as np

# Sample data generation (replace with actual data if available)
np.random.seed(0)
data_x = np.random.uniform(0, 700000, 300)  # Easting in meters
data_y = np.random.uniform(0, 1300000, 300)  # Northing in meters
color_category = np.random.choice(['red', 'blue'], 300)  # Color category

# Define plot area for the zoomed-in map in meters
zoom_xlim = (350000, 550000)
zoom_ylim = (300000, 450000)

# Create the main figure and axis
fig = plt.figure(figsize=(10, 15))

# Top plot - full map of the UK with OSGB coordinates
ax1 = fig.add_subplot(2, 1, 1, projection=ccrs.OSGB())
ax1.set_extent([0, 700000, 0, 1300000], crs=ccrs.OSGB())  # Full UK extent in British National Grid (OSGB) coordinates
ax1.coastlines(resolution='10m')
ax1.gridlines(draw_labels=False, linewidth=0.5, color='gray', alpha=0.7, linestyle='--')

# Scatter plot for the full map
for color in ['red', 'blue']:
    ax1.scatter(data_x[color_category == color], 
                data_y[color_category == color], 
                color=color, s=10, transform=ccrs.OSGB())

# Highlight zoomed-in area with a rectangle
rect = patches.Rectangle((zoom_xlim[0], zoom_ylim[0]), 
                         zoom_xlim[1] - zoom_xlim[0], 
                         zoom_ylim[1] - zoom_ylim[0],
                         linewidth=2, edgecolor='black', facecolor='none', 
                         transform=ccrs.OSGB())
ax1.add_patch(rect)

# Bottom plot - zoomed-in section
ax2 = fig.add_subplot(2, 1, 2, projection=ccrs.OSGB())
ax2.set_extent([zoom_xlim[0], zoom_xlim[1], zoom_ylim[0], zoom_ylim[1]], crs=ccrs.OSGB())
ax2.coastlines(resolution='10m')
ax2.gridlines(draw_labels=False, linewidth=0.5, color='gray', alpha=0.7, linestyle='--')

# Zoomed-in scatter plot
for color in ['red', 'blue']:
    ax2.scatter(data_x[color_category == color], 
                data_y[color_category == color], 
                color=color, s=10, transform=ccrs.OSGB())

# Manual labels for the grid
# Top plot labels
ax1.text(-0.1, 0.5, 'Northing (m)', va='bottom', ha='center', 
         rotation='vertical', rotation_mode='anchor', transform=ax1.transAxes, fontsize=12)
ax1.text(0.5, -0.1, 'Easting (m)', va='top', ha='center', transform=ax1.transAxes, fontsize=12)

# Bottom plot labels
ax2.text(-0.1, 0.5, 'Northing (m)', va='bottom', ha='center', 
         rotation='vertical', rotation_mode='anchor', transform=ax2.transAxes, fontsize=12)
ax2.text(0.5, -0.1, 'Easting (m)', va='top', ha='center', transform=ax2.transAxes, fontsize=12)

# Add orientation arrows for "N" in each plot
for ax in [ax1, ax2]:
    ax.text(0.95, 0.9, 'N', transform=ax.transAxes, ha='center', va='center', fontsize=12)
    ax.annotate('', xy=(0.95, 0.85), xytext=(0.95, 0.95),
                arrowprops=dict(arrowstyle="->", lw=1.5), transform=ax.transAxes)

# Add connecting lines from the top corners of the zoom box to the zoomed-in plot
# Define the top two corners of the zoom box in data coordinates
top_left = (zoom_xlim[0], zoom_ylim[1])
top_right = (zoom_xlim[1], zoom_ylim[1])

# Connect top left corner
con1 = ConnectionPatch(xyA=top_left, coordsA=ax1.transData,
                       xyB=(0, 1), coordsB=ax2.transAxes,
                       axesA=ax1, axesB=ax2,
                       color="black", lw=1.5)
fig.add_artist(con1)

# Connect top right corner
con2 = ConnectionPatch(xyA=top_right, coordsA=ax1.transData,
                       xyB=(1, 1), coordsB=ax2.transAxes,
                       axesA=ax1, axesB=ax2,
                       color="black", lw=1.5)
fig.add_artist(con2)

plt.show()
```
